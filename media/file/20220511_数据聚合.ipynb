```python
df = pd.DataFrame({'key1':['A','A','B','B','A'],
                  'data1':['one','two','one','two','one'],
                  'key2':[2,3,4,6,8],
                  'data2':[3,5,np.nan,3,7]})
df
```

```python
df.groupby('key2').mean()
```

```python
def range_data_group(arr):
    return arr.max()-arr.min()
```

```python
df2 = pd.DataFrame(np.arange(36).reshape((6,6)), columns=list('abcdef'))
df2
```

```python
df2['key'] = pd.Series(list('aaabbb'))
df2
```

```python
data_group = df2.groupby('key')
data_group
```

```python
[x for x in data_group]
```

```python
dict([x for x in data_group])
```

```python
dict([x for x in data_group])['a']
```

```python
type(dict([x for x in data_group])['a'])
```

```python
data_group.agg(sum)
```

```python
data_group.sum()
```

```python
data_group.agg(range_data_group)
```

```python
data_group.agg([range_data_group,sum,np.mean])
```

```python
data_group.agg([('极值',range_data_group),('和',sum),('平均值',np.mean)])
```

```python
data_group.agg({'a':sum,'b':np.mean,'c':range_data_group})
```
